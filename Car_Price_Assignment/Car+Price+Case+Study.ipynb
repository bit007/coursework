### Data Cleaning and Preparation


```python
# Converting all the string data into lowercase

str_cols = cars.select_dtypes(['object'])

cars[str_cols.columns] = str_cols.apply(lambda x: x.str.lower())
cars.head(5)
```

```python
# In carname we are getting both company and model. We only need company name in it.

cars['CarName'] = cars['CarName'].map(lambda x: x.split(" ", 1)[0])
cars['CarName'].unique()
```

```python
# We can see that there are some Names which might be misspelt. Fixing this issue.

cars['CarName'] = cars['CarName'].str.replace('alfa-romero', 'alfa_romeo')
cars['CarName'] = cars['CarName'].str.replace('maxda', 'mazda')
cars['CarName'] = cars['CarName'].str.replace('porcshce', 'porsche')
cars['CarName'] = cars['CarName'].str.replace('toyouta', 'toyota')
cars['CarName'] = cars['CarName'].str.replace('vokswagen', 'volkswagen')
cars['CarName'] = cars['CarName'].str.replace('vw', 'volkswagen')
cars['CarName'].unique()
```

```python
# Checking for unique values in all the columns
cars.nunique()
```

```python
# Converting categorical fields having only two unique values to binary

cars['fueltype'] = cars['fueltype'].map({ 'diesel': 0,'gas': 1})
cars['aspiration'] = cars['aspiration'].map({'turbo': 0, 'std': 1})
cars['doornumber'] = cars['doornumber'].map({'four': 0, 'two': 1})
cars['enginelocation'] = cars['enginelocation'].map({'rear': 0, 'front': 1})
```

```python
# Creating derived variables

# Volume of the car 
cars['car_volume'] = cars['carlength'] * cars['carwidth'] * cars['carheight']

# Area of the car 
cars['car_area'] = cars['carlength'] * cars['carwidth']

# Power to size ratio
cars['power_size_ratio'] = cars['horsepower'] / cars['enginesize']
```

```python
# Observing the unique values for variables having number of unique values less than 10 and are categorical

print(cars['carbody'].unique())
print(cars['drivewheel'].unique())
print(cars['enginetype'].unique())
print(cars['cylindernumber'].unique())
print(cars['fuelsystem'].unique())
```

```python
# We can convert cylindernumber to integers in order to reduce the number of categorical variables which would be increased
# once we create more dummmy variables

cars['cylindernumber'] = cars['cylindernumber'].map({'two': 2, 'three': 3, 'four': 4, 'five':5, 'six':6, 'eight':8, 'twelve':12})
```

```python
# Dropping the unique identifier

cars = cars.drop(['car_ID' ], axis=1)
cars.head(10)
```

```python
# Plotting the correlation plot

plt.figure(figsize = (16, 10))
sns.heatmap(cars.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

4. These variables can be dropped now or can be eliminated in RFE.

#### Creating Dummy Variables

Creating dummy variables for above columns CarName, carbody, drivewheel, enginetype, fuelsystem.
We would use get_dummies function to create a new dataframe and will keep drop_first=True to create n-1 dummies for n categories.

### Model Building 

```python
# Splitting the data in test and train datasets. Putting random_state = 100, so that it will always generate same splits.

from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(cars_df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# Rescaling features using mix max scaler

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_train.head(5)
```

```python
# Applying mix max scaling to all numerical variables.

num_vars = num_vars = ['symboling', 'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight','cylindernumber', 'enginesize', 'boreratio', 'stroke','compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price', 'car_volume', 'car_area', 'power_size_ratio']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.describe()
```

```python
# Creating x_train and y_train

y_train = df_train.pop('price')
X_train = df_train
```

```python
# Importing RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)  
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Columns to be used after RFE

X_train.columns[rfe.support_]
```

```python
# Columns which gets eliminated

cols_dropped = X_train.columns[~rfe.support_].tolist()
cols_dropped
```

#### Creating function to generate VIF

```python
# VIF Function takes input as dataframe and returns a dataframe showing VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor

def vif(df):
    vif_df = pd.DataFrame()
    X = df
    vif_df['Features'] = X.columns
    vif_df['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif_df['VIF'] = round(vif_df['VIF'], 2)
    vif_df = vif_df.sort_values(by = "VIF", ascending = False)
    return (vif_df)
```

#### Building model using statsmodel, for the detailed statistics

```python
lm_1 = sm.OLS(y_train,X_train_lm1).fit() 
print(lm_1.summary())
```

```python
# Getting the VIF of the model

vif(X_train1)
```
